```python
kb = 0
c = 1
ch = 'A'
l = list('')
disk = list('')

def add_ch(kb):
    for i in range(kb):
        l.append(ch)
    print(l)
    print('Память в Кб  Всего / Занято / Свободно:', 64, '/', len(l), '/', 64 - len(l))

def d_ch(del_ch):
    while (del_ch in l):
        l.remove(del_ch)
    print(l)
    print('Память в Кб  Всего / Занято / Свободно:', 64, '/', len(l), '/', 64 - len(l))
    
def re_ch(del_ch, l, disk):
    while (del_ch in l):
        disk.append(del_ch)
        l.remove(del_ch)
    print(l)
    print('Память в Кб  Всего / Занято / Свободно:', 64, '/', len(l), '/', 64 - len(l))
    print(disk)
    print('Память в Кб на диске:', len(disk))
    
def remo_ch(del_ch, disk, l):
    while (del_ch in disk):
        l.append(del_ch)
        disk.remove(del_ch)
    print(l)
    print('Память в Кб  Всего / Занято / Свободно:', 64, '/', len(l), '/', 64 - len(l))
    print(disk)
    print('Память в Кб на диске:', len(disk))


while (c != 0):
    print('0 - завершить работу программы  ', '1 - добавить задачу  ', '2 - удалить задачу')
    c = int(input())
   
    # добавить задачу
    if c == 1: 
        print('Введите размер процесса в Кб: ')
        kb = int(input())
        
        while (kb > 64 - len(l)): # нет свободного места
            del_ch = l[0]
            re_ch(del_ch, l, disk)
        
        add_ch(kb) # произвели запись
        
        for s in disk:
            while (64 - len(l) > disk.count(s) and len(disk) > 0): # можем перенести некоторые задачи с диска обратно
                remo_ch(s, disk, l)
                break
        
        ch = chr(ord(ch) + 1)

        
    # удалить задачу   
    elif c == 2:
        print('Введите задачу, которую нужно удалить:')
        del_ch = (input())
        d_ch(del_ch)
        
        for s in disk:
            while (64 - len(l) > disk.count(s) and len(disk) > 0): # можем перенести некоторые задачи с диска обратно
                remo_ch(s, disk, l)
                break
```
